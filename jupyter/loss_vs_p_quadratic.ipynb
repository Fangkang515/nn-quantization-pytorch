```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# plt.rcParams['figure.figsize'] = [12, 9]
import scipy.optimize as opt
import torch
import torch.nn as nn
import pickle
```

```python
data = pickle.load(open('../data/resnet18_WNoneA2_loss_vs_p.pkl', 'rb'))
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
font = {'weight' : 'normal', 'size': 16, 'weight': 'bold'}
matplotlib.rc('font', **font)

p = data['p']
loss = data['loss']

plt.scatter(p, loss, linewidth=4);
plt.ylabel('loss', fontsize=20)
plt.xlabel('p', fontsize=20);

xp = np.linspace(np.min(p), np.max(p), 100)
z = np.polyfit(p, loss, 2)
y = np.poly1d(z)
plt.plot(xp, y(xp), '--', linewidth=4, c='orange');

plt.legend(["L(p) = {:.2f}*p^2 + {:.2f}*p + {:.2f}".format(z[0], z[1], z[2])]);
```
